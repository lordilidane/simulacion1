```python
%matplotlib inline 
import matplotlib.pyplot as plt
import math

def  congruencial(x0):
    a = 1664525
    b = 1013904223
    m = 2**32
    xi = x0
    ui =0
    xi = (a*xi + b)%(m)
    ui = float(xi)/float(m)

    return xi,ui

def circunferencia(radio):
   
        xi = 43
        yi = 32
        randuXin =[]
        randuYin =[]
        randuXout =[]
        randuYout =[]
        for x in range(1000):
            xi,randuauxX = congruencial(xi)
            yi,randuauxY = congruencial(yi) 
            px = (2*radio*randuauxX-radio)
            py = (2*radio*randuauxY-radio) 
            distancia = math.sqrt(px**2+py**2)
            
            if (distancia < float(radio)) :
                randuXin.append(px)
                randuYin.append(py) 
            else:
                randuXout.append(px)
                randuYout.append(py)
        ax = plt.gca()
        ax.cla() 
        ax.set_xlim((-13, 13))
        ax.set_ylim((-10, 10))
        ax.plot(randuXin, randuYin, 'o')
        ax.plot(randuXout, randuYout, 'ro')
        circle1 = plt.Circle((0,0), 10, color='k',fill=False)
        ax.add_artist(circle1)
        
        plt.show()
        areaSimulada = (1000/(2*math.pi*(2*radio)**2))
        areaSimulada = areaSimulada * len(randuXin)
        #areaSimulada = math.pi
        print("area simulada")
        print(areaSimulada)
        print("area calculada matemáticamente")
        areaCalculada = math.pi*radio**2
        print(areaCalculada)
circunferencia(10)
```
